```python
# Loading the sql extension and connecting to the database
%load_ext sql

%sql sqlite://
```

```python
# Here we import the pandas library and CSV Library that we will use to open and manipulate our file

import csv 

import pandas as pd
```

```python
# We open the file 

with open('Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv', 'r') as f:
  UgandaCropYield=pd.read_csv(f, index_col=0, encoding='utf-8')


%sql DROP TABLE IF EXISTS UgandaCropYield;
%sql PERSIST UgandaCropYield;
```

```python
#Display a list of Sub Counties and their population and areas.

%%sql 

SELECT  "SUBCOUNTY_NAME", "POP", Area 

FROM UgandaCropYield ;
```

```python
# Sort the list of districts by total crop area (descending order).

%%sql 

SELECT "DISTRICT_NAME" 

FROM UgandaCropYield  

ORDER BY "Crop_Area_Ha" DESC;

```

```python
# Select only the Sub counties from the Moroto district, order them alphabetically 
# and show their production of sorghum.


%%sql 

SELECT "SUBCOUNTY_NAME"

FROM UgandaCropYield 

WHERE "DISTRICT_NAME" = "MOROTO"

ORDER BY "S_Prod_Tot" ASC;
```

```python
# Compute the total Maize production per District.

%%sql 

SELECT "DISTRICT_NAME", "M_Prod_Tot"

FROM UgandaCropYield
```

```python
# Compute the number of Sub counties where Maize is 
# produced and the total Maize production per District

%%sql 
SELECT "SUBCOUNTY_NAME", "M_Prod_Tot"

FROM UgandaCropYield 

WHERE "M_Prod_Tot" > 0;


```

```python
# Compute the overall Crop area in all
#  Sub counties where population is over 20000.

%%sql 

SELECT  "SUBCOUNTY_NAME", "Crop_Area_Ha"

FROM UgandaCropYield

WHERE "POP" > 20000;

```

```python
# Sort the Maize production in descending order by 
# Districts, only taking into account Sub counties 
# where Maize area is larger than Sorghum area, 
# and display the number of Sub counties
#  per district matching that criteria.


%%sql 

SELECT "DISTRICT_NAME", "M_Prod_Tot", "SUBCOUNTY_NAME"

FROM UgandaCropYield

WHERE "M_Area_Ha" > "S_Area_Ha" ;
```
